```python
#Basic One without Feature Eng

def featureEng(df):
    df=df.copy()
    df=df.drop(['Cabin','PassengerId','Name','Ticket'],axis=1)
    dtypeCov={
        'Pclass':'int8',
        'SibSp':'int8',
        'Parch':'int8',
    }
    df=df.astype(dtypeCov)
    return df
featureEngTrans=FunctionTransformer(featureEng)    
numericalCols=['Age','SibSp','Parch','Fare']
categoricalCols=['Pclass','Embarked','Sex']
numPipe=Pipeline(
    [
        ('imputer',SimpleImputer(strategy='mean')),
    ]
)
catePipe=Pipeline(
[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('encoder',OneHotEncoder(handle_unknown='ignore'))
]
)
preprocessor=ColumnTransformer(
    [
        ('num',numPipe,numericalCols),
        ('cat',catePipe,categoricalCols),
    ]    
)
pipeline=Pipeline(
    [
        ('featureEng',featureEngTrans),
        ('preprocessing',preprocessor),
        ('classifier',LogisticRegression(max_iter=1000))
    ]
)

pipeline.fit(X_train,y_train)
y_pred=pipeline.predict(X_test)
acc=accuracy_score(y_test,y_pred)
cv_score=cross_val_score(pipeline,X,y,cv=10,scoring='accuracy')

print(acc)
print(cv_score.mean())
```
